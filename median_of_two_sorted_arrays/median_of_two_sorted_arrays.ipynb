## 4. Algorithm design

1. Compute `n=len(nums1)` and `m=len(nums2)`.
    * Compare the medians of the two lists. Maybe they are already the same.
    * Check if one of the two lists is empty.


2. Setup a binary search algorithm.
    
    * Check if `n+m` is odd or even.
    * The initial stepsize is set with `min(n,m)//4`. 
    * Setup a function to check if the median was found.


3. Start iteration
    
    * Check where to move.
    * move the median positions in oposite directions.
    * cut step_size by half until you reach 1.


```python
def EvaluateMedianGuess(nums1,nums2,mid1,mid2):
    # This function evaluates the median guess pair (mid1,mid2).
    # The return spectrum is ('found','left','right').
    # 'found': indicates (mid1,mid2) is a good match.
    # 'left': mid1 moves left and mid2 moves right.
    # 'right': the inverse of left.
    
    # Get the current values of the common median guess from mid1 and mid2
    median1 = nums1[mid1]
    median2 = nums2[mid2]
    
    # Necessary condition for median1 and median2 to be in the direct vicinity of the common median
    mid1_fits = (median1 >= nums2[mid2-1]) and (median1 <= nums2[mid2+1])
    mid2_fits = (median2 >= nums1[mid1-1]) and (median2 <= nums1[mid1+1])
    
    # If true. Check for the actuall median.
    if mid1_fits and mid2_fits:
        #Left and right are the count of elements to the left and right of median1 and median2
        left = mid1+mid2
        right = len(nums1)+len(nums2)-2-left
        
        # Combine the two medians into a single, sorted array
        if median1>=median2:
            mid = [median2, median1]
        else:
            mid = [median1, median2]
            
        # make sure that both the left and right side of the mid list are equal in length
        while left!=right:
            
            # running indices to take elements from the left or right of the mid list
            jl1 = 1
            jl2 = 1
            jr1 = 1
            jr2 = 1
            
            if left<right:
                right += -1
                #take the smallest element of the right side and add it to the mid list
                if nums1[mid1+jr1]>=nums2[mid2+jr2]:
                    mid.append(nums2[mid2+jr2])
                    jr2 += 1
                else:
                    mid.append(nums1[mid1+jr1])
                    jr1 += 1
            else:
                left += -1
                # take the largest element on the left side and add it to the mid list
                if nums1[mid1-jl1]<=nums2[mid2-jl2]:
                    mid.insert(0,nums2[mid2-jl2])
                    jl2 += 1
                else:
                    mid.insert(0,nums1[mid1-jl1])
                    jl1 += 1
        
        l = len(mid)
        if l%2==0:
            return "found", (mid[l//2]+mid[l//2-1])/2
        else:
            return "found", mid[l//2]
        
    # if mid1 and mid2 are not candidates for median positions, determine where to move relative to mid1/nums1.
    if median1>median2:
        return "left", None
    else:
        return "right", None
```

```python
def findMedianSortedArrays(nums1, nums2):
    # get the length
    n = len(nums1)
    m = len(nums2)
    
    # Checking special cases
    # Check if one of the two lists is empty 
    if n==0:
        if m%2==1:
            return nums2[m//2]
        else:
            return (nums2[m//2]+nums2[m//2-1])/2
        
    if m==0:
        if n%2==1:
            return nums1[n//2]
        else:
            return (nums1[n//2]+nums1[n//2-1])/2
    
    # Check if the two lists are sorted relative to each other.
    if nums1[-1]<nums2[0]:
        nums = nums1 + nums2
        n = len(nums)
        if n%2==1:
            return nums[n//2]
        else:
            return (nums[n//2]+nums[n//2-1])/2
    
    if nums1[0]>nums2[-1]:
        nums = nums2 + nums1
        n = len(nums)
        if n%2==1:
            return nums[n//2]
        else:
            return (nums[n//2]+nums[n//2-1])/2
        
    # get the middle 
    mid1 = n//2
    mid2 = m//2
    
    # check medians
    if n%2==1:
        median1 = nums1[mid1]
    else:
        median1 = (nums1[mid1]+nums1[mid1-1])/2
    if m%2==1:
        median2 = nums2[mid2]
    else:
        median2 = (nums2[mid2]+nums2[mid2-1])/2
        
    # if you got lucky this is the median.
    if median1==median2:
        return median1
    
    # get the stepsize. It needs to be the minimum of the two medians
    step_size = max(min(mid1,mid2)//4,1)
    
    while True:
        # compare the median locations
        move, median = EvaluateMedianGuess(nums1,nums2,mid1,mid2)
        
        # evaluate the comparison.
        if move=="found":
            return median
        elif move=="right":
            mid1 += step_size
            mid2 += -step_size
        elif move=="left":
            mid1 += -step_size
            mid2 += step_size
            
        # Ensuring that both mid1 and mid2 are valid.
        if mid1>=n:
            mid1=n-1
        elif mid1<0:
            mid1=0
        if mid2>=m:
            mid2=m-1
        elif mid2<0:
            mid2=0
            
        # Reduce the step size for the next step.
        step_size = max(step_size//2,1)
```
